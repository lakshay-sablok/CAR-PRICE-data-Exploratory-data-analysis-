# **FUELTYPE** :- GAS & DIESEL

```python
fueltype=top_t['fueltype'].value_counts().head(10)
fueltype
```

```python
fig, ax = plt.subplots(figsize = (20,8))
ax=sns.barplot(y=top_t['price'], x=top_t['fueltype'], palette='magma')
plt.title('Top 70 Cars FuelType prices',color='black',fontsize=30)
plt.xticks(rotation=0, fontsize=30)
plt.yticks(fontsize=30)

ax.set_xlabel('\nFUELTYPE\n', fontsize=30, color='black')
ax.set_ylabel('\nPRICE\n', fontsize=30, color='black')

ax.grid(color='black', linewidth=1, axis='y', alpha=.2)

plt.show
```
